# Simple Concat by Rows

```python
import pandas as pd
import numpy as np
```

```python
# assigning table1
student_data1 = pd.DataFrame({
    'student_id':['S1','S2','S3','S4','S5'],
    'name':['Danniella Fenton', 'Ryder Storey', 'Bryce Jensen', 'Ed Bernal', 'Kwame Morin'],
    'marks':[200, 210, 190, 222, 199]
})

# assigning table2
student_data2 = pd.DataFrame({
    'student_id':['S6', 'S7', 'S8', 'S9', 'S10'],
    'name': ['Scarlette Fisher', 'Carla Williamson', 'Dante Morse', 'Kaiser William', 'Madeeha Preston'], 
    'marks': [201, 200, 198, 219, 201]
})

# concat by rows
pd.concat([student_data1, student_data2], ignore_index=True)
```

```python
# checking duplicate names 1
for i in list(student_data1['name'].values):
    print("# {}".format(i))
    if i in list(student_data2['name'].values):
        print('Name duplicated')
    else:
        print('Name not duplicated')
```

```python
# checking duplicate names 2
list(set(student_data1['name'].values) & set(student_data2['name'].values))
```

```python
# checking duplicate marks 2
list(set(student_data1['marks'].values) & set(student_data2['marks'].values))
```

# Simple Concat by Columns

```python
# concat by columns
pd.concat([student_data1, student_data2], axis=1)
```

# Appending by row (series & dicts) - single

```python
# new row1
row1 = pd.Series(student_data2.iloc[0])
row1
```

```python
# new row2
s7 = pd.Series(['S7','hupla',230], index=['student_id','name','marks'])
```

```python
# appending row1
student_data1.append(row1)
```

```python
# appending row2 (appending series must have ignore_index=True)
student_data1.append(s7, ignore_index=True)
```

# Concating created series

```python
ser1 = pd.Series([0,1,2,3,4,5])
ser2 = pd.Series([2,3,4,5,6,7])
ser3 = pd.Series([3,4,5,6,7,8])

pd.concat([ser1,ser2,ser3], axis=1, keys=['colA','colB','colC'])
```

# Concat Rows & Joining to another Table

```python
# recreating variables
student_data1 = pd.DataFrame({
        'student_id': ['S1', 'S2', 'S3', 'S4', 'S5'],
         'name': ['Danniella Fenton', 'Ryder Storey', 'Bryce Jensen', 'Ed Bernal', 'Kwame Morin'], 
        'marks': [200, 210, 190, 222, 199]
})

student_data2 = pd.DataFrame({
        'student_id': ['S4', 'S5', 'S6', 'S7', 'S8'],
        'name': ['Scarlette Fisher', 'Carla Williamson', 'Dante Morse', 'Kaiser William', 'Madeeha Preston'], 
        'marks': [201, 200, 198, 219, 201]
})

exam_data = pd.DataFrame({
        'student_id': ['S1', 'S2', 'S3', 'S4', 'S5', 'S7', 'S8', 'S9', 'S10', 'S11', 'S12', 'S13'],
        'exam_id': [23, 45, 12, 67, 21, 55, 33, 14, 56, 83, 88, 12]
})
```

```python
# concatenating student data1  & data2
A = pd.concat([student_data1, student_data2])
A
```

```python
# merging (there's duplicate student id hence duplicate exam id)
A.merge(exam_data, how='inner', left_on='student_id', right_on='student_id')
```

# Inner/Outer Merging by Student Id

```python
# inner merging
pd.merge(student_data1, student_data2, on='student_id', how='inner')
```

```python
# outer merging
pd.merge(student_data1, student_data2, on='student_id', how='outer')
```

```python
# left merging
pd.merge(student_data1, student_data2, on='student_id', how='left')
```

```python
# right merging
pd.merge(student_data1, student_data2, on='student_id', how='right')
```

# Merging on multiple keys

```python
# assigning labels
data1 = pd.DataFrame({
    'key1':['K0','K0','K1','K2'],
    'key2':['K0','K1','K0','K1'],
    'P':['P0','P1','P2','P3'],
    'Q':['Q0','Q1','Q2','Q3']
})

data2 = pd.DataFrame({
    'key1':['K0','K1','K1','K2'],
    'key2':['K0','K0','K0','K0'],
    'R':['R0','R1','R2','R3'],
    'S':['S0','S1','S2','S3']
})
```

```python
data1.merge(data2, how='left', on=['key1','key2'])
```

```python
data1.merge(data2, how='right', on=['key1','key2'])
```

# On 'inner' == without on Merging

```python
data1.merge(data2, on=['key1','key2'])
```

```python
data1.merge(data2,how='inner', on=['key1','key2'])
```

# Combining Columns & filling the values from each other

```python
df1 = pd.DataFrame({
    'A':[None, 1, None],
    'B':['a','b',None]
})

df2 = pd.DataFrame({
    'A':[20, None, 100],
    'B':[None,None,'c']
})

df1.combine_first(df2)
```
